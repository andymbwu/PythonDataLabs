Count the number of occurences of the words in the file. Make a functions that gets the list of words generated in the part 1 of the problem as input and outputs a dictionary containing the number of occurences of each word.  

```python
def count_words(word_list):
    word_frequencies_dict = {}
    for word in word_list:
        if word in word_frequencies_dict:
            word_frequencies_dict[word] += 1
        else:
            word_frequencies_dict[word] = 1
    return word_frequencies_dict
print(count_words(word_list))
```

Using python's built-in `defaultdict` can help you write a shorter code. Rewrite above function using `defaultdict`. Take a look at python's documentation for more information. 

```python
from collections import defaultdict
def count_words(word_list):
    word_frequencies_dict = defaultdict(int)
    for word in word_list:
        word_frequencies_dict[word] += 1
    return word_frequencies_dict
    
word_frequencies_dict = count_words(word_list)
#for word, frequency in word_frequencies_dict.items():
    #print(word,frequency)
    
```

## Part 2 (20 marks)

Find the maximum, and [median](https://en.wikipedia.org/wiki/Median) of the word frequencies (i.e. the number of occurences). To find the median you need to sort the words and their respective frequecies by the frequency. Use python's `sorted` function. You will need to create a list of tuples based on the dictionary. Each tuple should contain the word and its frequency. Use `NamedTuple`s to improve your code's readablity. Use the `key` attribute of the `sorted` function to define what the element the sort should be based on.  

```python
from collections import namedtuple
def sorted_word_frequencies(word_frequencies_dict):
    """Ouput a sorted list of tuples as described above."""
    # Your solution goes here
    tupleList = list()
    Frequency = namedtuple('Frequency', 'word frequency')
    for word, frequency in word_frequencies_dict.items():
        tupleList.append(Frequency(word, frequency))
    sortedTuples = sorted(tupleList, key = lambda tup:tup[1], reverse = True)
    return sortedTuples
word_frequency_pairs = sorted_word_frequencies(word_frequencies_dict)
#for word in word_frequency_pairs:
    #print(word_frequency_pairs) 
    
    
def find_median(word_frequency_pairs):
    """Find the median of word frequencies"""
    lengthList = len(word_frequency_pairs)
    if(lengthList % 2 == 0):
        index = (lengthList-1) // 2
        return (word_frequency_pairs[index][1]+word_frequency_pairs[index+1][1])/2
    else:
        index = (lengthList-1) // 2 + 1
        return word_frequency_pairs[index][1]
median = find_median(word_frequency_pairs)
#print(median)
    
def find_maximum(word_frequency_pairs):
    """Find the maximum occurence. You will only need the maximum, so you don't have to sort all the frequencies at this point"""
    maximum = word_frequency_pairs[0][1]
    return maximum
maximum_occurence = find_maximum(word_frequency_pairs)
#print(word_frequency_pairs[0][0] , ":" , maximum_occurence, "times")
```

```python
def normalize_frequencies(word_frequency_pairs):
    tempDict = {a:b for a,b in word_frequency_pairs}
    for key,values in tempDict.items():
        tempDict[key] = values/maximum_occurence
        normFreq = [(a,b) for a,b in tempDict.items()]
    return normFreq
normalizedFrequencies = normalize_frequencies(word_frequency_pairs)
#print(normalizedFrequencies)

    
    
```

## Part 3 (20 marks)
We want to check if the normalized frequencies have any corelation to their ranks. If such correlation exists, the Zipf's law states that it is linear in a log-log space. Take the list of normalized frequencies and create a list of log normalized data points for each word. Create a list of tuples containing the word itself, `log(frequency)` and `log(rank)`. Since the list is already sorted `rank` will be `index_in_list + 1` (Because ranks start at 1 and list indexes start at 0, you need that `+ 1`). 

Calculate the `pearson correlation coefficient` on this data points. The result is expected to be close to `-1`. Define appropriate functions for the the statistical calculations as neccessary. Alternatively, you can reuse `pearsonr` function from [scipy package](https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.pearsonr.html). Both options are acceptable.  

```python
import math
from scipy.stats.stats import pearsonr
def create_log_normalized_points(normalizedFrequencies):
    tempDict = {a:b for a,b in normalizedFrequencies}
    for key,values in tempDict.items():
        tempDict[key] = math.log(values)
    tempList = list(range(1,len(tempDict)+1))
    newtempList = [math.log(x,10) for x in tempList]
    frequencyRankTuple = [(a,b,c) for ((a,b),c) in zip(tempDict.items(),newtempList)]
    
    return frequencyRankTuple
data_points = create_log_normalized_points(normalizedFrequencies)
#print(data_points)


def pcc(data_points):
    x = [values[1] for values in data_points]
    y = [values[2] for values in data_points]
    coeff = pearsonr(x,y)
    return coeff
pearsonCoeff = pcc(data_points)
print(pearsonCoeff)
    
```

# The email template delimma (50 marks)

We'd like to have a program that creates email messages based on some predefined templates. We break this problem down to a couple of smaller problems and merge them to get the final program.

## Part 1 (5 marks)
Let's start by defining a function that gets a template and a value to fill in the template. The template is a string that contains some substrings that look like `[%some name%]`. We call them placeholders. What we want to do is to find the placeholders and replace them with appropriate values. Let's start simple and assume the only placeholder that we have is `[%name%]`. Complete the function below:

```python
def replace_name_in_template(template, name):
    # Your solution goes here
    replacement = '[%name%]'
    newTemplate = template.replace(replacement, name)
    return newTemplate
        

actual_result = replace_name_in_template("Hello [%name%]! How are you doing today [%name%]?", "John")
expected_result = 'Hello John! How are you doing today John?'
if actual_result == expected_result:
    print('Good!')
else:
    print('Wrong, "', actual_result, '" != "', expected_result)
```

## Part 2 (20 marks)

Now we're going to tweak it by supporting more placeholders, not just `[%name%]`. Complete the function below to get a `template` as a string and a `context` dictionary. The `context` provides the placeholder names and the values to be replaced. See the example for more clarification.

```python
def parse_template(template, context):
    # Your solution goes here
    for placeholder, replacement in context.items():
        template = template.replace("[%"+placeholder+"%]", replacement)
    return template
    

template = """Hi [%name%]!

Hope you're doing well! Can we meet on [%weekday%] [%time%] to have a brief discussion about [%subject%]?

Best,
[%my_name%]"""
context = {
    "name": "John",
    "weekday": "Thursday",
    "time": "8am",
    "subject": "some serious stuff",
    "my_name": "Lindsey",
}

actual_result = parse_template(template, context)
expected_result = """Hi John!

Hope you're doing well! Can we meet on Thursday 8am to have a brief discussion about some serious stuff?

Best,
Lindsey"""
if actual_result == expected_result:
    print('Good!')
else:
    print('Wrong, "', actual_result, '" != "', expected_result)
```

**Note that the context can have arbitrary keys. "name", "weekday", "time", ... are just some examples.**
Also, Note that just like the first part, `replace_name_in_template`, the placeholders can appear more than once. You must substitude all occurances of the placeholder with the provided value from the context dictionary.
