## The Crust

* Create a class named `Crust`, it must have:
    * a `size` data attribute which is an integer (size in inches) and is fed to the class via its constructor. Check the validity of the input.
    * It must inherit both `HasPriceMixin` and `DietaryRestrictionsMixin` and should override `get_price`, and `is_vegeterian` methods properly.
    
* Create classes for each "kind" of available crusts (3 classes in total). These must be sub-classes of the `Crust` class. In each sub-class, you need to have a **class attribute** that holds the available sizes and their respective prices (Using a dictionary would be the best option). 

Here's the table of prices for each size and type of crusts.

| Price of each dough  | 10"   | 12"   | 14"   | 16"   |
|----------------------|-------|-------|-------|-------|
| Original Hand Tossed | 5.00  | 7.00  | 9.00  | 11.00 |
| Crunchy Thin Crust   | 4.50  | 6.50  | 8.50  |       |
| Handmade Pan         | 7.50  |       |       |       |

```python
class HasPriceMixin:
    def get_price(self) -> float:
        raise NotImplementedError()
        
class DietaryRestrictionsMixin:
    def is_vegeterian(self) -> bool:
        raise NotImplementedError()

class Crust(HasPriceMixin,DietaryRestrictionsMixin):
    def __init__(self,size):
        self.size = size
    def get_price(self,size) -> float:
        return self.price
    def is_vegetarian(self,size) -> bool:
        return False
class OriginalHandTossed(Crust):
    dictSizes = {10:5.00,12:7.00,14:9.00,16:11.00}
    def __init__(self,size):
        if size in self.__class__.dictSizes:
            self.size = size
        else:
            raise ValueError()
    def get_price(self) -> float:
        self.price = self.__class__.dictSizes.get(self.size)
        return self.price
    
class CrunchyThinCrust(Crust):
    dictSizes = {10:4.50,12:6.50,14:8.50}
    def __init__(self,size):
        if size in self.__class__.dictSizes:
            self.size = size
        else:
            raise ValueError()
    def get_price(self) -> float:
        self.price = self.__class__.dictSizes.get(self.size)
        return self.price
        
class HandmadePan(Crust):
    dictSizes = {10:7.50}
    def __init__(self,size):
        if size in self.__class__.dictSizes:
            self.size = size
        else:
            raise ValueError()
    def get_price(self) -> float:
        self.price = self.__class__.dictSizes.get(self.size)
        return self.price
```

Just like the `amount`, it will be fed into your class via the constructor and your code must check for the validity of the input. Use `"normal"` as the default value.

* Create a `MeatTopping` class, which inherits `Topping`. Override `is_vegeterian` function.

**Topping classes:**

**Price changes based on the area:**

**Price changes based on the amount of the topping:**

| Amount | Price Coefficient |

In the end you will end up with a hierarchy of 10 classes (`Topping`, `MeatTopping`, `NonMeatTopping`, and 7 classes for each kind of topping)


```python
class Topping(HasPriceMixin,DietaryRestrictionsMixin):
    name:str = ''
    base_price:float = 0.0
    dictToppings = {'Pepperoni':0.0,'Chicken':2.0,'Beef':2.5,'Cheddar':0.75,'Onion':0.0,'Basil':0.5,'Mushroom':0.0}
    dictArea = {'whole':1.0,'left':0.5,'right':0.5}
    dictAmount = {'light':0.7,'normal':1,'extra':1.3,'double':2,'triple':3}
        
    def __init__(self,area:str = 'whole',amount:str = 'normal'):
        if area not in self.__class__.dictArea:
            raise ValueError()
        else:
            self.area = area
        if amount not in self.__class__.dictAmount:
            raise ValueError()
        else:
            self.amount = amount
        
class MeatTopping(Topping):
    def is_vegeterian(self) -> bool:
        return False
    
class NonMeatTopping(Topping):
    def is_vegeterian(self) -> bool:
        return True

class Pepperoni(MeatTopping):
    def get_price(self) -> float:
        self.base_price = Topping.dictToppings.get(self.__class__.__name__)
        areaCoeff:float = self.__class__.dictArea.get(self.area)
        amountCoeff:float = self.__class__.dictAmount.get(self.amount)
        return areaCoeff*amountCoeff*self.base_price
class Chicken(MeatTopping):
    def get_price(self) -> float:
        self.base_price = Topping.dictToppings.get(self.__class__.__name__)
        areaCoeff:float = self.__class__.dictArea.get(self.area)
        amountCoeff:float = self.__class__.dictAmount.get(self.amount)
        return areaCoeff*amountCoeff*self.base_price
class Beef(MeatTopping):
    def get_price(self) -> float:
        self.base_price = Topping.dictToppings.get(self.__class__.__name__)
        areaCoeff:float = self.__class__.dictArea.get(self.area)
        amountCoeff:float = self.__class__.dictAmount.get(self.amount)
        return areaCoeff*amountCoeff*self.base_price
class Cheddar(NonMeatTopping):
    def get_price(self) -> float:
        self.base_price = Topping.dictToppings.get(self.__class__.__name__)
        areaCoeff:float = self.__class__.dictArea.get(self.area)
        amountCoeff:float = self.__class__.dictAmount.get(self.amount)
        return areaCoeff*amountCoeff*self.base_price
class Onion(NonMeatTopping):
    def get_price(self) -> float:
        self.base_price = Topping.dictToppings.get(self.__class__.__name__)
        areaCoeff:float = self.__class__.dictArea.get(self.area)
        amountCoeff:float = self.__class__.dictAmount.get(self.amount)
        return areaCoeff*amountCoeff*self.base_price
class Basil(NonMeatTopping):
    def get_price(self) -> float:
        self.base_price = Topping.dictToppings.get(self.__class__.__name__)
        areaCoeff:float = self.__class__.dictArea.get(self.area)
        amountCoeff:float = self.__class__.dictAmount.get(self.amount)
        return areaCoeff*amountCoeff*self.base_price
class Mushroom(NonMeatTopping):
    def get_price(self) -> float:
        self.base_price = Topping.dictToppings.get(self.__class__.__name__)
        areaCoeff:float = self.__class__.dictArea.get(self.area)
        amountCoeff:float = self.__class__.dictAmount.get(self.amount)
        return areaCoeff*amountCoeff*self.base_price
```

## The sauce
Create a class named `Sauce` that has the following attributes:

* an `amount` data attribute: It accepts `"normal"`, `"light"`, and `"extra"` values. Like all other classes check the validity of the input and use `"normal"` as the default value. 
* Inherit `HasPriceMixin` and `DietaryRestrictionsMixin` and override their methods appropriately. The price is affected by the amount and the base price. The change in price based on the value of `amount` is just like the toppings.

Create two `Sauce` subclasses with following base prices:

| Sauce Name   | Price |
|--------------|-------|
| PizzaSauce   | 1.00  |
| AlfredoSauce | 1.75  |


```python
class Sauce(HasPriceMixin,DietaryRestrictionsMixin):
    base_price:float = 0.0
    dictSauce = {'PizzaSauce':1.0,'AlfredoSauce':1.75}
    dictAmount = {'none':0.0,'light':0.7,'normal':1.0,'extra':1.3}
    
    def __init__(self,amount:str = 'normal'):
        if amount not in ('none','normal','light','extra'):
            raise ValueError(amount)
        else:
            self.amount = amount
    def get_price(self) -> float:
        self.base_price = 0.0
        amountCoeff = self.dictAmount.get('none')
        return amountCoeff*self.base_price

class PizzaSauce(Sauce):
    def get_price(self) -> float:
        self.base_price = Sauce.dictSauce.get(self.__class__.__name__)
        amountCoeff = Sauce.dictAmount.get(self.amount)
        return amountCoeff*self.base_price
class AlfredoSauce(Sauce):
    def get_price(self) -> float:
        self.base_price = Sauce.dictSauce.get(self.__class__.__name__)
        amountCoeff = Sauce.dictAmount.get(self.amount)
        return amountCoeff*self.base_price
        
    
        
```


## The pizza itself

Create a class named `Pizza` that has the following attributes:

* a `cheese` data attribute: The amount of cheese on the pizza. It can accept `None` (no cheese), `"normal"`, and `"extra"` as value. It will be provided to the class through the constructor. Check for input validity. If no value was provided use `"normal"` as the default value.
* a `crust` data attribute: An instance of one of the `Crust` sub-classes. Check for the validity of input (the input must be a `Crust` subclass). This input obviously can't be `None` and there is no default value. It must be present while creating the pizza class.
* a `sauce` data attribute: An instance of one of the "Sauce" classes. Check for the validity of input. This input can be `None`, meaning the pizza will not have any sauce. default value should be `None`
* a `toppings` data attribute: A `set` of toppings. 
* an `add_topping` method: A method that with help of `*args` gets a list of toppings to add them to the pizza. Since we're using a `set` to store the toppings, one topping can't be added twice to the pizza.
* Inherit `HasPriceMixin` and `DietaryRestrictionsMixin` and override their methods appropriately. Calculate the price by adding up all the prices of the toppings, sauce and the crust. Whether it's vegeterian is similarly determined if all the ingredients have the desired property.


**Cheese Price:**

| Amount    | Price  |
|-----------|--------|
| No cheese | 0.00   |
| Light     | 1.00   |
| Normal    | 2.50   |
| Extra     | 4.00   |

```python
class Pizza(HasPriceMixin,DietaryRestrictionsMixin):
    
    cheesePrices = {None:0.00,'light':1.00,'normal':2.50,'extra':4.00}
    
    def __init__(self, crust:object,cheese:str = 'normal',sauce:object = Sauce('none'), toppings:set = None):
        if cheese not in self.__class__.cheesePrices:
            raise ValueError()
        else:
            self.cheese = cheese
        if isinstance(crust,Crust):
            self.crust = crust
        else:
            raise ValueError()
        if isinstance(sauce,Sauce):
            self.sauce = sauce
        else:
            raise ValueError()
        self.toppings = set()
        if toppings is not None:
            for topping in toppings:
                self.toppings.add(topping)
    def add_topping(self, *args):
        for topping in args:
            self.toppings.add(topping)
    def is_vegeterian(self) -> bool:
        isVegetarian = True
        for topping in self.toppings:
            if topping.is_vegeterian() == False:
                isVegetarian = False
        return isVegetarian
    
    def get_price(self) -> float:
        toppingPrice = 0.0
        for topping in self.toppings:
            toppingPrice += topping.get_price() 
        totalCheesePrice = 0.0
        totalCheesePrice += self.__class__.cheesePrices.get(self.cheese)
        saucePrice = 0.0
        saucePrice += self.sauce.get_price()
        
        #print(self.crust.get_price(),toppingPrice,saucePrice,totalCheesePrice)
        totalPrice = self.crust.get_price() + toppingPrice + saucePrice + totalCheesePrice
        print("Total pizza price is: $", totalPrice)
        return totalPrice
            
```
